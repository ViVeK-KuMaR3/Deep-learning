```python
doc=nlp(text)
```

```python
tokens=[word for word in doc]
```

```python
tokens
```

```python
len(tokens)
```

```python
tokens=[word for word in tokens if str(word) not in punctuation]
```

```python
len(tokens)
```

```python
tokens
```

```python
tokens=[word for word in tokens if str(word) not in stopwords]
```

```python
len(tokens)
```

```python
tokens
```

```python
tokens=[str(word).lower() for word in tokens if str(word).isalpha()]
```

```python
tokens
```

```python
from collections import Counter
```

```python
word_distribution=dict(Counter(tokens))
```

```python
word_distribution
```

```python
max_freq=max(word_distribution.values())
```

```python
max_freq
```

```python
for word in word_distribution.keys():
    word_distribution[word]=round(word_distribution[word]/max_freq,2)
```

```python
word_distribution
```

```python
sentence_tokens=[sent for sent in doc.sents]
```

```python
sentence_tokens
```

```python
sentence_scores={}
for sent in sentence_tokens:
    for word in sent:
        if word.text.lower() in word_distribution.keys():
            if sent not in sentence_scores.keys():
                sentence_scores[sent]=word_distribution[word.text.lower()]
            else:
                sentence_scores[sent]+=word_distribution[word.text.lower()]
                
```

```python
sentence_scores
```

```python
from heapq import nlargest
```

```python
select_length = int(len(sentence_tokens)*0.3)
```

```python
select_length
```

```python
summary =  nlargest(select_length,sentence_scores,key=sentence_scores.get)
```

```python
summary
```

```python
final_summary=[word.text for word in summary]
```
